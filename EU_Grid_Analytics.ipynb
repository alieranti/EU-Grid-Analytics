# Load and Clean the Data

```python
import pandas as pd

# Define the path to your file on Google Drive
file_path = '/content/drive/MyDrive/vertice_price_df_anonymized.csv'

# Load the data
df = pd.read_csv(file_path)

# Clean it by dropping rows that are missing price data
df = df.dropna(subset=['price'])

df
```

# Calculate Intra-Country Congestion

```python
# Group by country and calculate the stats
country_stats = df.groupby('country').agg(
    Avg_Price=('price', 'mean'),
    Min_Price=('price', 'min'),
    Max_Price=('price', 'max'),
    Node_Count=('name', 'count')
).reset_index()

# Calculate the "Price Spread" (Max minus Min)
country_stats['Price_Spread'] = country_stats['Max_Price'] - country_stats['Min_Price']

# Sort the table to see which country has the worst grid congestion at the top
country_stats = country_stats.sort_values(by='Price_Spread', ascending=False)

country_stats
```

# Spain

```python
# Filter out only the nodes in Spain (ES)
spain_nodes = df[df['country'] == 'ES'].copy()

# Sort them from most expensive to least expensive
spain_nodes = spain_nodes.sort_values(by='price', ascending=False)

spain_nodes
```

# Export the Clean Data Back to Drive

```python
# Define paths to save the new files directly to your Google Drive
clean_nodal_path = '/content/drive/MyDrive/cleaned_nodal_prices.csv'
country_stats_path = '/content/drive/MyDrive/country_congestion_stats.csv'
spain_targets_path = '/content/drive/MyDrive/spain_siting_targets.csv'

# Export the dataframes to CSV
df.to_csv(clean_nodal_path, index=False)
country_stats.to_csv(country_stats_path, index=False)
spain_nodes.to_csv(spain_targets_path, index=False)

print("Success! Your files are now saved in your Google Drive.")
```
